```python
import pandas as pd
```

```python
wine_quality = pd.read_csv("/content/drive/MyDrive/WineQT.csv", index_col='Id')
```

```python
wine_quality.head(5)
```

To convert all the values in the quality attribute to
0 (bad) if the value is less than or equal to 6 and to 1 (good) otherwise, I am going to use a lambda function along with apply to get the desired output

```python
# the lambda function definition

def convert_quality(x):
  if x <= 6:
    return 0
  else:
    return 1
```

```python
wine_quality['quality'] = wine_quality['quality'].apply(convert_quality)
```

```python
wine_quality.head(5)
```

The quality values are converted as desired

Now we have to  Normalize all the other attributes between 0 and 1 by min-max scaling.
<br>
We will be using MinMaxScaler from sklearn.preprocessing

```python
from sklearn.preprocessing import MinMaxScaler
```

We will exclude quality column from the normalisation process.

```python
exclude_columns = ['quality']
```

Now we will create the base for the columns which we want to include in the normalising process

```python
included_columns = wine_quality.columns.difference(exclude_columns)
included_columns
```

initializing the minmaxscaler

```python
scaler = MinMaxScaler()
```

Normalising the included_columns

```python
wine_quality[included_columns] = scaler.fit_transform(
    wine_quality[included_columns]
    )
```
